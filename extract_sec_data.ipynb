```python
import urllib.request
import os
base_url = 'https://www.sec.gov/files/dera/data/financial-statement-data-sets/'
#save_to directory  eg.'/home/xx/xxxx/'
#from_y from year eg. 2009, to_y eg. 2020, the effective dates are from 2009 to 2020.
def download_financial_statement_data(from_y, to_y, save_to):
    print('Beginning file download with urllib2...')
    for y in range(from_y, to_y + 1):
        #get quaters
        for q in range(1, 5):
            param = str(y)+'q'+str(q)
            url = base_url + param +'.zip'
        print(url)
        #change this line to your folder
        urllib.request.urlretrieve(url, os.path.join(save_to, param))
print('the finacial report index is downloaded!')
```

```python
#!/usr/bin/python3
#this code is use to extract the finacial report metadata.
from zipfile import ZipFile
import pandas as pd
import urllib
import os 
from urllib.error import URLError, HTTPError

year = 2020
quater = 'q4'
base_dir = '/home/weiwei/delta_intern/'
zip_path = base_dir+'/'+ str(year) + quater
archive_url = 'https://www.sec.gov/Archives/edgar/data/'
finacial_report = 'Financial_Report.xlsx'

#unzip downloaded finacial statement data files
#
#path_to_zip_file e.g. /home/weiwei/delta_intern/2020q4.zip
def get_report_indices(base_path, year, save_reports_to):
    with ZipFile(base_path, 'r') as zip_ref:
        sub_txt = zip_ref.open('sub.txt')
        df = pd.read_csv(sub_txt, sep="\t")
        df = df[df['fp'].notna()]
        print('the total records shape is:')
        print(df.shape)
        company_list = df[['adsh', 'cik','fp']].values
        count = 0
        for company in company_list:
            adsh = company[0].replace('-','')
            cik = company[1]
            fp = company[2]
            url = archive_url + str(cik) + '/' + adsh + '/' + finacial_report
            save_name = adsh + '_' + str(cik) + '_' + fp + '.xlsx'
            if count < 5:
                print(url)
            try:
                urllib.request.urlretrieve(url, os.path.join(save_reports_to, save_name))
            except HTTPError as e:
                print('Error code: ', e.code)
                print(url)
            count += 1
            
save_reports_to = os.path.join(base_dir, str(year))
if not os.path.exists(save_reports_to):
    os.makedirs(save_reports_to)
    
get_report_indices(base_path = zip_path, year = 2020, save_reports_to = save_reports_to)
    

    
```
